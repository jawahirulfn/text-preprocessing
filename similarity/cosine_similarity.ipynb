```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_excel("data_similarity.xlsx")
texts = df['news_title'].tolist()
texts
```

```python
# Inisialisasi ruang vektor
tfidf_vectorizer = TfidfVectorizer()

# Transform standard text menjadi vektor
tfidf_standard_matrix = tfidf_vectorizer.fit_transform(texts)

# Define a new input phrase
input_phrase = "Partisipasi aktif Prabowo dalam Pilpres \
menarik perhatian publik dan menciptakan dinamika politik \
yang menarik dalam konteks demokrasi Indonesia."

# Transform the input phrase
tfidf_input = tfidf_vectorizer.transform([input_phrase])

# Menghitung Cosine similarity antara input phrase dan standard phrase
cosine_sim_scores = cosine_similarity(tfidf_input, tfidf_standard_matrix)

index = 1
for ph, score in zip(tfidf_standard_matrix, cosine_sim_scores[0]):
    print(f"Similarity data {index}: {score:.4f}") 
    index +=1
```
